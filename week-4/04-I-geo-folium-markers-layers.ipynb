````python
#Â tarde = evening
#Â maÃ±ana = morning
#Â noche = night


#Â lesividad = moderada, grave & leve

#Â festividad
    #Â festivo: holiday
    # laborable: working day
    # vÃ­spera: day before a holiday
````

We are going to iterate over all the rows of the dataframe looking for what is inside each row in the column.
We will put an icon to the accidents based on the "holiday" column
But we will have to save the settings and add the Marker at the end and all this inside a loop ðŸ¤­ðŸ¤¯
- First the tooltip config
- Then the little icon depending on the holiday
- Finally add each marker

````python
#https://datos.madrid.es/portal/site/egob/menuitem.9e1e2f6404558187cf35cf3584f1a5a0/?vgnextoid=374512b9ace9f310VgnVCM100000171f5a0aRCRD&vgnextchannel=374512b9ace9f310VgnVCM100000171f5a0aRCRD&vgnextfmt=default
````

Is that the best way to visualize this? We have created a map with infinite markers...but is there another way to visualize this information a little better?
When we have a lot of grouped data ðŸ—º

````python
mad_lat = 40.4169376
mad_lon = -3.706187217

madrid_map = Map(location=[mad_lat, mad_lon], zoom_start = 15)
madrid_map
````

````python
df.sample()
````

````python
for index, row in df.iterrows():
    
    #1. MARKER without icon
    district = {"location": [row["lat"], row["lon"]], "tooltip": row["distrito"]}
    
    #2. Icon
    if row["festividad"] == "Laborable":        
        icon = Icon (
            color="blue",
            opacity = 0.6,
            prefix = "fa",
            icon="briefcase",
            icon_color = "black"
        )
    elif row["festividad"] == "Vispera":
        icon = Icon (
            color="red",
            opacity = 0.6,
            prefix = "fa",
            icon="beer",
            icon_color = "yellow"
        )
    elif row["festividad"] == "Festivo":
        icon = Icon (
            color="darkpurple",
            opacity = 0.6,
            prefix = "fa",
            icon="bed",
            icon_color = "white"
        )
    else:
        icon = Icon (
            color="blue",
            opacity = 0.6,
            prefix = "fa",
            icon="question",
            icon_color = "brown",
            icon_size=(14, 14)
        )
    #3. Marker
    new_marker = Marker(**district, icon = icon, radius = 2)
    
    #4. Add the Marker
    new_marker.add_to(madrid_map)
````

````python
madrid_map
````

##  Morning bikers will be safer  ðŸš´

Let's check how many categories there are in the time column, to make a heat map. First we make a map with the morning schedule, and then we add the afternoon schedule with the Layer Control.
Here we go ðŸ”¥

### We create the basic map to which we then add the layers

````python
mad_lon
mad_lat
````

````python
madrid_map_2 = Map (location = [mad_lat, mad_lon], zoom_start = 15)
````

````python
madrid_map_2
````

````python
df["horario"].unique()
````

### Subset the dataframe per layer

````python
morning = df[df["horario"] == "MaÃ±ana"]
morning.sample()
````

### We create the FeatureGroup

````python
morning_group = folium.FeatureGroup(name = "Morning")
````

### Create the HeatMap layer and add it to the FeatureGroup

````python

HeatMap(data = morning[["lon", "lat"]]).add_to(morning_group)
````

### Add the FeatureGroup with heatmap and name to the map

````python
morning_group.add_to(madrid_map_2)
````

````python
madrid_map_2
````

###Â Adding the evenings

````python
evening = df[df["horario"] == "Tarde"]
evening_group = folium.FeatureGroup(name=f"Evening {evening.shape[0]}")

HeatMap(data = evening[["lat", "lon"]]).add_to(evening_group)
evening_group.add_to(madrid_map_2)
madrid_map_2
````

### Night layer

````python
night = df[df["horario"] == "Noche"]
night_group = folium.FeatureGroup(name=f"Night: {night.shape[0]}")

HeatMap(data = night[["lat", "lon"]]).add_to(night_group)
night_group.add_to(madrid_map_2)
madrid_map_2
````

````python
folium.LayerControl(collapsed=False, position="topleft").add_to(madrid_map_2)
````

````python
madrid_map_2
````

### We can save our maps
