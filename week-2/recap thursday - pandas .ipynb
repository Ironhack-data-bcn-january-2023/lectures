- apply: apply a function to a column in a dataframe
        - `lambda x: x`
        - `def my_function (n)` 
        - one or more args:
            - if one: 
                - `lambda cell: cell.upper()`
                - `df["column"].apply(my_function)`
            - if more:
                - `lambda row: row["column"]`
                - `def my_function (row, n)`: `df.apply(my_function(row["column"], n))`
         - to a Series | a df: this will affect as the input will be different
         
                
    - similar to `map()`
    - either `inplace=True` or you save into new/same column

- `loc` & `iloc`: they locate things
    - `loc`: Labels
    - `iloc`: Indexes
    - `iloc[2:4, 0:-5]`: rows 2, 3 & columns 0, 1, 2 .. -5
        - like coordinates
        - df.iloc[2:4][0:1]


- AGGREGATION: groupby & pivot table
    - groupby: 
        - df[[col1, col2]].groupby("col2").agg("max") -> it can take a dictionary
        - groupby: df[[col1, col2, col3]].groupby("col2, col3").agg({"col2": "max", "col3":"mean"}) -> it can take a dictionary

        - aggregate
            - groupby: df[[col1, col2, col3]].groupby("col2, col3").mean()

    - pd.cross_tab: two categorical variables: frequency: how many times does a category happen in another one
  

- duplicate values
    - duplicated(): how many times a row has been repeated: if 3 rows are equal: 1 row has been repeated twice
        - drop_duplicates()
  
- fillna
    - fillna: mean
    - fillna: categorial ("string")
    - ignore them
    - drop them 

- COMBINATION: concat, merge & join

    - concat: side to side
    - merge: mixes looking for the common column
    - join: common index
    
    - types of join/union/merge/whatever
        - left: we prioritze the data on the left & other matches on the right
        - right: same
        - inner: conservative approach, only whatever is in both
        - outer: all of them except for the common ones
