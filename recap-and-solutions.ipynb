```python
# 2nd approach: .count() - David


#1. What we have: two things: one iterable(tuple) & another iterable (list)
tup3 = ('I', 'r', 'o', 'n', 'h', 'a', 'c', 'k', "a")
letters = ["a", "b", "c", "d", "e"]

#2. What we want to get/return: frequency of something
    #"for each element one iterable: how many times it happens in the other iterable"
    # 1. something (common elementas) & 2. number (how many times those elements happen)

lst_aux=[] #lst_aux=["a", "c"]


# c scenario
for i in letters:
    if i in tup3: #false, condition is not met
        lst_aux.append(i)
        print (f"{i} is True it's on the tuple tup3")
    else:
        print(f"instead {i} is not" )
        
print(lst_aux)

# lst_aux is a list with the common elements

# c
for j in lst_aux:
    if j in tup3:
        print(f"{j} is repeated", tup3.count(j), "times")
```
